```python
import pymongo, os, pandas as pd
from dotenv import load_dotenv
load_dotenv()

client = pymongo.MongoClient(os.getenv("MONGO_URI"))
db = client["quotesdb"]
collection = db["quotes"]
print(f"Connected! {collection.count_documents({})} quotes in DB")
```

```python
# Pull all documents, drop the MongoDB _id field
df = pd.DataFrame(list(collection.find({}, {"_id": 0})))
print(df.shape)
df.head()
```

```python
df["author"].value_counts().head(10)
```

```python
# MongoDB query — equivalent to: SELECT * WHERE tags CONTAINS 'love'
love_quotes = pd.DataFrame(list(
    collection.find({"tags": "love"}, {"_id": 0})
))
print(f"{len(love_quotes)} quotes tagged 'love'")
love_quotes[["author", "text"]]
```

```python
tag_counts = (df["tags"]
    .explode()          # one row per tag
    .value_counts()
    .head(10))
tag_counts.plot(kind="barh", title="Top 10 tags")
```
